```python
from rdflib import Graph


g = Graph()
g.parse("../attrattori.rdf")
print(len(g))
```

```python
query_all = f"""
SELECT *
WHERE {{
    ?type ?pred ?obj .
}}
LIMIT 20
"""

SPARQL_PREFIXES = """
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX geo: <http://www.w3.org/2003/01/geo/wgs84_pos#>
PREFIX dcterms: <http://purl.org/dc/elements/1.1/>
PREFIX umb: <http://dati.regione.umbria.it/tourism/ontology/> 
PREFIX lgdo: <http://linkedgeodata.org/ontology/>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
"""

query1 = """
SELECT ?titolo ?descrizione ?lat ?long ?immagine_copertina
WHERE {
  ?s umb:id '4130375'^^xsd:string ;
     rdfs:label ?titolo ;
     umb:testo ?descrizione ;
     geo:lat ?lat ;
     geo:long ?long .
  
  OPTIONAL { ?s umb:immagine_copertina ?immagine_copertina . }
  
  FILTER(LANG(?titolo) = "it" && LANG(?descrizione) = "it")
}"""

query2 = f"""
SELECT ?id ?titolo ?provenienza
WHERE {{
  ?s rdf:type lgdo:Attraction ;  # Sostituire dinamicamente con la category opportuna
     umb:id ?id ;
     umb:titolo_testo ?titolo ;
     dcterms:provenance ?provenienza .
     
  FILTER(LANG(?titolo) = "it")
}}
LIMIT 20
"""



results = g.query(SPARQL_PREFIXES + query1)

print(str(results))
results.serialize(destination="results.txt", format="txt")
results.serialize(destination="results.csv", format="csv")

# for row in results:
#     print(row[n] for n in range(len(row)))
```
